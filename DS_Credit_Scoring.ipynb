## Truncamiento de Valores atpicos

Segn los histogramas tenemos valores inusuales para nuestros datos. Son valores que auqneu pueden ser o ms altos o ms bajos que la mayoria de datos que tenemos, son muy pocos regiistros asi que no represnetan un valor alto para nuestro modelo por ello vamos a tratarlos haciendo un truncamiento de valores, para este caso usaremos los percentiles 0.05 y 0.99.

Lo que haremos es que lo que sea menor a 0.05 se tomado com igual a 0.05 para juntar esos datos, y lo que sea mayor a 0.99 lo tomaremos como si fuese 0.99 y asi tambien traeremos esos datos a este rango de valores.

Asi nuestros datos seran ms representativos y funcionales para nuestro modelo.

```python
#primero calcularemos los percentiles 95 y 99 de algunas columna para decidir qu porcentaje de datos vamos a truncar
columnas = ['expenses','income', 'assets', 'debt',	'amount',	'price']
limits = {}

#calculamos ambos percentiles
for col in columnas:
  low = dataset[col].quantile(0.05)
  high= dataset[col].quantile(0.99)  # Percentiles a calcular
  limits[col] = (low, high)          #guardamos en el diccionario

pprint (limits)
```

```python
# Aplicamos el truncamiento
for col, (low, high) in limits.items():               # para cada columna segun limite inferior y superior dado en el diccioinario
    dataset[col] = np.clip(dataset[col], low, high)   #aplicamos el truncamiento

```

```python
fig, axes = plt.subplots(2,3, figsize=(16, 8))         #creamos la grfica 2 col, 3 filas, especificamos tamao
ax_flat=axes.flatten()                                 #aplanamos para facilitar la iteraccin, aunque al ser de 2
                                                      #columnas es bidimensioal, lo tratara unidimensional para recorrerlo ms facilmente

for i, col in enumerate(columnas):                           #iteramos
    dataset[col].hist(ax=ax_flat[i], bins=30, edgecolor='black')   #grficamos
    ax_flat[i].set_title(col)

plt.tight_layout()                                           #ajustamos espacios
plt.show()
```

## Calculamos y grficamos el riesgo

Ahora calcularemos el riesgo.

*   Riesgo global:\
Es la proporcin de clientes que incumplen (default = 1) en todo el dataset.

*   Riesgo por categora:\
Es la proporcin de clientes que incumplen dentro de cada categora de una columna categrica.


*   Ratio de riesgo:\
Es el riesgo de una categora dividido por el riesgo global.
Indica cunto ms (o menos) probable es que una categora incumpla en comparacin con el promedio global.



```python
#seleccionamos columnas categricas
cat_cols = dataset.select_dtypes(include=['object']).columns

#calclulamos el riesgo global, o sea la media
#la propocin de clientes q incumplen en todos los datos q tenemos
global_risk = dataset['default'].mean()

#funcin para calcular por categoria y el ratio
def risk_analysis(column):

  #calcular por categoria
  #recordemos que groupby agrupa por los valores unicos de cada columna y asi obtenemos la categoria
  grouped = dataset.groupby(column)['default'].mean()

  #calculamos el ratio
  #Es el riesgo de una categora dividido por el riesgo global
  risk_ratio = grouped / global_risk

  #devolvemos un df con los datos generados (redondeados y ordenados)
  return (pd.DataFrame({'Risk': grouped.round(4), 'Risk Ratio': risk_ratio.round(4)}) #redondeamos valores
                       .sort_values(by='Risk', ascending=False))                      #organizamos descendientemente

for col in cat_cols:
  #imprimimos el dato de riesgo individual
  print(f'analiss de riesgo para la columna: {col}')

  #ejecutamos la funcin
  print(risk_analysis(col))
  print("\n" + "="*50 + "\n")  # Separador entre columnas
```

Teniendo ya los datos de cada categoria y el riesgo que representa tener esa categoria en nuestros datos, grficamos

```python
def plot_risk(column):
    grouped = risk_analysis(column)['Risk']
    fig, ax = plt.subplots(figsize=(10, 6))

    # Usar una paleta especfica
    colors = sns.color_palette("Set2", len(grouped))  # Ajustamos el nmero de colores al nmero de categoras

    # Crear el grfico
    bars = ax.bar(grouped.index, grouped.values, color=colors)

    # Aadir detalles al grfico
    ax.set_title(f'Riesgo de default por categora: {column}', fontsize=16)
    ax.set_xlabel(column, fontsize=12)
    ax.set_ylabel('Riesgo (Proporcin de default)', fontsize=12)
    ax.set_ylim(0, max(grouped.values) * 1.2)  # Ajustar el lmite superior del eje y

    # Calcular y aadir porcentajes y anlisis
    total = sum(grouped.values)
    for i, bar in enumerate(bars):
        height = bar.get_height()
        percentage = (height / total) * 100
        ax.text(bar.get_x() + bar.get_width()/2, height,
                f'{percentage:.1f}%\n{grouped.index[i]}: {height:.3f}',
                ha='center', va='bottom', fontweight='bold')

    # Aadir anlisis general
    max_category = grouped.idxmax()
    min_category = grouped.idxmin()
    analysis_text = (f"Anlisis:\n"
                     f"Mayor riesgo: {max_category} ({grouped[max_category]:.3f})\n"
                     f"Menor riesgo: {min_category} ({grouped[min_category]:.3f})")
    ax.text(0.95, 0.95, analysis_text, transform=ax.transAxes,
            ha='right', va='top', bbox=dict(facecolor='white', alpha=0.8))

    plt.tight_layout()
    plt.show()

# Iterar sobre cada columna categrica y graficar el riesgo
for col in cat_cols:
    plot_risk(col)
```

Balancearemos el dato para ello usaremos... \
***Over_samplin***
usaremos ***SMOTE***

Este va a generar lineas con la clase menor, en este casso 1, y va a crear lineas con pequeas variaciones en algunas valores para que el resultado sea el mismo pero las variables X sean diferentes. Esto aportar valor al modelo y se utiliza cuando tenemos una clase menor, solo sirve si la diferencia de proporcin no es muy extrema. Porque tampoco podemos crear demasiados datos sintticos ya que seria demasiado falso y all no aportariamos nada bueno al modelo.

```python
#importamos la biblioteca necesaria
from imblearn.over_sampling import SMOTE
```

```python
#dividimos nuestros datos
#en x eliminamos u obviamos default ya que es la variable a predecir
x = dataset_con_dummies.drop('default', axis=1)
y = dataset_con_dummies['default']           #y contendr nuestra variable a predecir
```

Para ello usaremos algunas mtricas que nos permitiran responder

El modelo est haciendo predicciones correctas?

Cuntos aciertos y errores tiene el modelo?

Es lo suficientemente bueno para usarse en un entorno real?

Si tu objetivo es minimizar los falsos negativos, enfcate en recall.

Si tu objetivo es minimizar los falsos positivos, enfcate en precisin.

Si buscas un equilibrio, usa F1-Score.

Si quieres una evaluacin general, usa ROC-AUC o el reporte de clasificacin.

Matriz de confusin
Muestra los aciertos y errores del modelo desglosados por clase.

Nos permite ver:

True Positives (TP): Predicciones correctas de la clase positiva.

True Negatives (TN): Predicciones correctas de la clase negativa.

False Positives (FP): Predicciones incorrectas de la clase positiva (falsos positivos).

False Negatives (FN): Predicciones incorrectas de la clase negativa (falsos negativos).

En credit scoring, los falsos negativos (FN) son especialmente costosos, ya que representan clientes que incumplirn pero el modelo no los detect.

Recall (Sensibilidad)
Mide la capacidad del modelo para identificar correctamente los casos positivos.

Frmula: TP / (TP + FN).

Es til cuando es importante minimizar los falsos negativos (por ejemplo, en diagnsticos mdicos).

Importancia: En credit scoring, es crucial maximizar el recall para minimizar los falsos negativos (clientes que incumplirn pero no fueron detectados).

Precisin (Precision)
Mide la proporcin de predicciones positivas que son correctas.

Frmula: TP / (TP + FP).

Es til cuando es importante minimizar los falsos positivos (por ejemplo, en deteccin de spam).

Importancia: Te ayuda a evaluar cuntos de los clientes identificados como incumplidores realmente lo son. Un alto precision significa menos falsos positivos.

 F1-Score
Es una combinacin de precisin y recall.

Frmula: 2 * (Precision * Recall) / (Precision + Recall).

Es til cuando buscas un equilibrio entre precisin y recall.

Importancia: Es til cuando buscas un equilibrio entre identificar correctamente los casos de default (recall) y minimizar los falsos positivos (precision).

ROC-AUC (rea bajo la curva ROC)
Mide la capacidad del modelo para distinguir entre las clases default y no default.

La curva ROC grafica la tasa de verdaderos positivos (Recall) contra la tasa de falsos positivos (FPR).

AUC: Un valor cercano a 1 indica un modelo perfecto, mientras que un valor cercano a 0.5 indica un modelo aleatorio.

Importancia: Es especialmente til cuando las clases estn desequilibradas.

### Evaluacin del modelo

```python
def evaluar_modelo(Y_test, predic_logic, y_pred_proba=None, titulo="Evaluacin del Modelo"):
    """
    Evala el rendimiento de un modelo de clasificacin.

    Parmetros:
    - Y_test: Etiquetas reales.
    - predic_logic: Etiquetas predichas.
    - titulo: Ttulo para la grfica de la matriz de confusin.
    """
    # Matriz de confusin
    conf_matrix = confusion_matrix(Y_test, predic_logic)
    conf_matrix_df = pd.DataFrame(conf_matrix,
                                  index=["Real 0", "Real 1"],
                                  columns=["Predicho 0", "Predicho 1"])

    # Grfica de la matriz de confusin
    plt.figure(figsize=(6, 4))
    sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicho')
    plt.ylabel('Real')
    plt.title(titulo)
    plt.show()

    # Mtricas individuales
    precision = precision_score(Y_test, predic_logic)
    recall = recall_score(Y_test, predic_logic)
    f1 = f1_score(Y_test, predic_logic)

    print(f'Precisin (Precision): {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1-Score: {f1:.2f}')

    # Reporte de clasificacin
    print("Reporte de clasificacin:")
    print(classification_report(Y_test, predic_logic))

      # Curva ROC (si se proporcionan probabilidades)
    if y_pred_proba is not None:
        fpr, tpr, thresholds = roc_curve(Y_test, y_pred_proba)
        roc_auc = roc_auc_score(Y_test, y_pred_proba)

         # Imprimir el valor del AUC
        print(f'AUC (rea bajo la curva ROC): {roc_auc:.2f}')

        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Curva ROC')
        plt.legend(loc="lower right")
        plt.show()
```

```python
# Obtener probabilidades predichas
y_pred_proba = mod_logic.predict_proba(X_test)[:, 1]  # Probabilidades de la clase positiva

#llamamos a la funcin
# Evaluar el modelo con probabilidades y curva ROC
evaluar_modelo(Y_test, predic_logic, y_pred_proba=y_pred_proba, titulo="Evaluacin Inicial del Modelo")
```

### Umbral optimo

```python
# Calcular fpr, tpr, y thresholds
fpr, tpr, thresholds = roc_curve(Y_test, y_pred_proba)

# Encontrar el umbral ptimo (donde TPR - FPR es mximo)
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]
print(f'Umbral ptimo: {optimal_threshold}')

# Aplicar el umbral ptimo para clasificar
predic_logic2 = (y_pred_proba >= optimal_threshold).astype(int)

# Evaluar el modelo con el nuevo umbral
evaluar_modelo(Y_test, predic_logic2, y_pred_proba=y_pred_proba, titulo="Evaluacin del Modelo con Umbral Optimizado")
```
